```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\Shant\OneDrive\Documents\vehicle performnace analysis\vehicle_performance.csv')
```

```python
df.horsepower.replace('?',np.nan,inplace=True)
```

```python
df['horsepower']=df['horsepower'].astype(float)
```

```python
df.horsepower.fillna(df['horsepower'].median(),inplace=True)
```

```python
#model building
```

```python
#independent features and dependent features
bx=df.drop(['name','Kilometer_per_liter'],axis=1)
```

```python
y=df['Kilometer_per_liter']
```

```python
from sklearn.model_selection import train_test_split
```

```python
bx_train,bx_test,y_train,y_test= train_test_split(bx,y,train_size=0.8,random_state= 42)
```

```python
from sklearn.linear_model import SGDRegressor
```
